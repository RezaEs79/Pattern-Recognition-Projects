```python
import numpy as np


x = np.array([[-7, 1/22],
                 [-5, 2/22],
                 [-4, 3/22],
                 [-3, 3/22],
                 [-2, 2/22],
                 [0, 3/44],
                 [2, 2/22],
                 [3, 3/22],
                 [4, 3/22],
                 [5, 2/22],
                 [7, 1/22]])

# Calculate the absolute values of x
abs_x = np.abs(x[:, 0])

# Check if the centers abs(x) are greater than 7
if abs_x.any() > 7:
  print("At least one center abs(x) is greater than 7.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([[-7, 1/22],
                 [-5, 2/22],
                 [-4, 3/22],
                 [-3, 3/22],
                 [-2, 2/22],
                 [0, 3/44],
                 [2, 2/22],
                 [3, 3/22],
                 [4, 3/22],
                 [5, 2/22],
                 [7, 1/22]])

x = data[:, 0]
values1 = data[:, 1]

# Values between -20 and -7
x1 = np.linspace(-20, -7, 1000)

# Values between 7 and 20
x2 = np.linspace(7, 20, 1000)

# Concatenate the arrays
x_combined = np.concatenate((x1, x2))

# Additional function
mask = np.abs(x_combined) > 7
values2 = np.zeros_like(x_combined)
values2[mask] = 1 / (11 * np.abs(5 - np.abs(x_combined[mask])))

# Plot both functions
plt.plot(x, values1, label='Original Function')
plt.plot(x_combined, values2, label='Additional Function')
plt.xlabel('x')
plt.ylabel('Values')
plt.title('Plot of Two Functions')
plt.legend()
plt.show()
```
